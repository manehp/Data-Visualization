###### 1. Create a Boxplot of Lung Cancer Deaths Distribution. (Python & R) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/Users/Administrator/Desktop/lung_cancer_prediction_dataset.csv")
df2 = pd.read_csv("/Users/Administrator/Desktop/global_air_pollution_dataset.csv")
```

```python
sns.boxplot(y=df1['Annual_Lung_Cancer_Deaths'], color= "magenta")
plt.title("Boxplot of Lung Cancer Deaths Distribution")
plt.ylabel("Number of Deaths")
plt.show()
```

###### 2. Create a Histogram of PM2.5 AQI Values.

```python
sns.histplot(df2['PM2.5_AQI_Value'], bins=30, color="Coral")
plt.title("Histogram of PM2.5 AQI Values")
plt.xlabel("PM2.5 AQI Value")
plt.ylabel("Frequency")
plt.show()
```

###### 3. Create a Density Plot of the Lung Cancer Mortality Rate.

```python
sns.kdeplot(df1['Mortality_Rate'], fill=True, color="hotpink", alpha=1)
plt.title("Density Plot of Lung Cancer Mortality Rate")
plt.xlabel("Mortality Rate")
plt.ylabel("Density")
plt.show()
```
